```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

IMPORT DATASETS FROM THE RESOURCES

```python
liver=pd.read_csv('/content/sample_data/Liver_data.csv')
liver.fillna(0)
```

IDENTIFY THE SHAPE OF ROWS AND COLUMNS

```python
liver.shape
```

IDENTIFY THE ROWS AND COLUMNS

```python
liver.columns
```

CHECK THE MISSING VALUE OF THE DATASET

```python
liver.isna().sum()
```

REMOVE THE ADDITIONAL VALUES FROM THE DATASET

```python
liver.dropna(inplace=True)
liver.isna().sum()
```

IDENTIFY THE SHAPE OF THE REMOVED DATASET

```python
liver.shape
```

IDENTIFY THE DATATYPE OF THE EACH COLUMNS

```python
liver.dtypes
```

WE FIND THE VARIOUS GENDER OF THE DATASET

```python
liver['Gender'].value_counts()
```

SHOW  THE GRAPH IN GENDER

```python
plt.figure(figsize=(6,3))
plt.title("GRAPH BETWEEN GENDER")
sns.countplot(x='Gender',data=liver,color='red')
plt.show()
```

TO MEASURE THE AMOUNT OF SUBSTANCE IN THE BLOOD TEST

```python
liver['Total_Bilirubin'].value_counts()
```

TO IDENTIFY THE BILIRUBIN USING GRAPH

```python
plt.figure(figsize=(6,12))
plt.title('AMOUNT OF SUBSTANCE IN THE BLOOD LEVEL')
sns.histplot(x='Total_Bilirubin',data=liver,kde=True,color='orange')
plt.show()
```

TO IDENTIFY THE DIRECT_BILIRUBIN USING GRAPH

```python
plt.figure(figsize=(6,12))
plt.title("BILIRUBIN IS CHANGED INTO THE FOR OF YOUR BODY CAN RID")
sns.countplot(x='Direct_Bilirubin',data=liver)
plt.show()
```

TO MEASURE THE AMOUNT OF ALP IN YOUR BODY

```python
plt.figure(figsize=(6,12))
plt.title("alkaine_phosphate")
sns.countplot(x='Alkaline_Phosphotase',data=liver)
plt.show()
```

BLOOD TEST CHECK FOR LIVER DAMAGE

```python
plt.figure(figsize=(6,12))
plt.title('ALAMINE AMINOTRANSFERASE')
sns.countplot(x='Alamine_Aminotransferase',data=liver,palette='winter')
plt.show()
```

TO IDENTIFY THE ENZYMES IN THE LIVER OR OTHER ORGAN IN THE BODY

```python
plt.figure(figsize=(6,12))
plt.title('ENZYMES IN THE LIVER')
sns.countplot(x='Aspartate_Aminotransferase',data=liver,palette='rocket')
plt.show()
```

TO CALCULATE THE TOTAL PROTEIN PRESENT THE LIVER

```python
plt.figure(figsize=(12,24))
plt.title('TOTAL PROTEIN PRESENT IN THE LIVER')
sns.countplot(x='Total_Protiens',data=liver,palette='winter')
plt.show()
```

TO IDENTIFY THE CIRCULATION PROTEIN PRESENT IN THE PLASMA

```python
plt.figure(figsize=(6,12))
plt.title('ALBUMIN')
sns.countplot(x='Albumin',data=liver,palette='rocket')
plt.show()
```

TO CALCULATE THE ALBUMIN AND GLOBULIN RATIO

```python
plt.figure(figsize=(6,12))
plt.title('Albumin_and_Globulin_Ratio')
sns.countplot(x='Albumin_and_Globulin_Ratio',data=liver,palette='winter')
plt.show()
```

TO FIND THE OUTPUT OF THE DATASET

```python
plt.figure(figsize=(6,12))
plt.title("Total outputs dataset")
sns.countplot(x='Dataset',data=liver,palette='rocket')
plt.show()
```
